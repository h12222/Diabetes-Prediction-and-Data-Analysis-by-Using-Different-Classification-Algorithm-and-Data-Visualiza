# Diabetes Prediction and Data Analysis by Using Different Classification Algorithm and Data Visualization

```python
#import the libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Read the dataset
df = pd.read_csv("diabetes analysis.csv")
df
```

```python
dfs=df.drop(columns=["Unnamed: 2","Unnamed: 5","Unnamed: 7","Unnamed: 9","Unnamed: 11","Unnamed: 12"])
dfs
```

```python
#Take all the independent Variable
X=dfs.iloc[:,:-1]
X
```

```python
# Y as a Target Variable
y=df.iloc[:,-1]
y
```

```python
# Train and Test the Data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=1)
```

```python
X_train.fillna(0,inplace=True)
```

```python
y_train.fillna(0,inplace=True)
X_test.fillna(0,inplace=True)
y_test.fillna(0,inplace=True)
```

# Random Forest

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
accuracy_rate = []

# Will take some time
for i in range(1,24):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,dfs,dfs['Diabetes'],cv=10)
    accuracy_rate.append(score.mean())
```

```python
error_rate = []

# Will take some time
for i in range(1,24):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,dfs,dfs['Diabetes'],cv=10)
    error_rate.append(1-score.mean())
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,24),error_rate,color='blue', linestyle='dashed', marker='o',
        markerfacecolor='red', markersize=10)
#plt.plot(range(1,24),accuracy_rate,color='blue', linestyle='dashed', marker='o',
        # markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
print('WITH K=5')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
```

```python
knn.score(X_test,y_test)
```
